```python
import torch
import cv2
import os

# Check if video file exists
video_path = '/content/demo.gif'
if not os.path.isfile(video_path):
    raise FileNotFoundError(f"Video file not found: {video_path}")

# Load YOLOv5 model (custom or pretrained)
model = torch.hub.load('ultralytics/yolov5', 'custom', path='/content/yolov5/runs/train/exp/weights/best.pt', trust_repo=True)

class_names = {
    0: 'bus_stop',
    1: 'do_not_enter',
    2: 'do_not_stop',
    3: 'do_not_turn_l',
    4: 'do_not_turn_r',
    5: 'do_not_u_turn',
    6: 'enter_left_lane',
    7: 'green_light',
    8: 'left_right_lane',
    9: 'no_parking',
    10: 'parking',
    11: 'ped_crossing',
    12: 'ped_zebra_cross',
    13: 'railway_crossing',
    14: 'red_light',
    15: 'stop',
    16: 't_intersection_l',
    17: 'traffic_light',
    18: 'u_turn',
    19: 'warning',
    20: 'yellow_light'
}

# Function to detect objects in a frame
def detect_objects_in_frame(frame, confidence_threshold=0.3):
    # Convert the frame from BGR to RGB (YOLO expects RGB)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Perform inference on the frame
    results = model(rgb_frame)

    # Filter out results based on the confidence threshold
    detections = []
    for *box, conf, cls in results.xyxy[0].tolist():
        if conf > confidence_threshold:
            detections.append({'bbox': box, 'confidence': conf, 'class': int(cls)})
    return detections

# Open the input video
cap = cv2.VideoCapture(video_path)

# Get video properties for VideoWriter
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# Set up VideoWriter to save the output video
output_path = '/content/output11.mp4'
output = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'mp4v'), fps, (width, height))

# Process the video frame by frame
while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Detect road signs in the current frame
    detections = detect_objects_in_frame(frame)

    # Draw bounding boxes and labels on detected objects
    for detection in detections:
        x1, y1, x2, y2 = map(int, detection['bbox'])
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        class_name = class_names[detection['class']] if detection['class'] < len(class_names) else "Unknown"
        label = f"{class_name}, Conf: {detection['confidence']:.2f}"
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Write the processed frame to the output video
    output.write(frame)

# Release video and resources
cap.release()
output.release()
cv2.destroyAllWindows()

print(f"Video with detections saved to {output_path}")
```
